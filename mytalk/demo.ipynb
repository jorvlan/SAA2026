# Install OMSS

```python
%matplotlib inline
import omss
from omss import Ruletype, AttributeType, Rule, create_matrix, plot_matrices, ruleset
```

OMSS works with elements which are placed in a 3 x 3 matrix. Elements' appearance is determined by their atributes. Specific instances of these attributes may change within a row according to a logical pattern. This logical pattern is determined by the user with customized rules or predifined standard rulesets.

# A basic first example

```python
rules = {
    'BigShape': [       
        Rule(Ruletype.DISTRIBUTE_THREE, AttributeType.SHAPE),
        Rule(Ruletype.CONSTANT, AttributeType.ANGLE),
        Rule(Ruletype.CONSTANT, AttributeType.COLOR),
        Rule(Ruletype.CONSTANT, AttributeType.NUMBER),
        Rule(Ruletype.FULL_CONSTANT, AttributeType.SIZE, value = 'medium')]}

solution_matrix, problem_matrix = create_matrix(rules, save = False)

plot_matrices(solution_matrix, problem_matrix)
```

# Swap custom rules for predifined rulesets

```python
solution_matrix, problem_matrix = create_matrix(ruleset.easy, save = False)

plot_matrices(solution_matrix, problem_matrix)
```

# Increase difficulty (easy to moderate)
# `ruleset.easy` -> `ruleset.moderate`

```python
solution_matrix, problem_matrix = create_matrix(ruleset.moderate, save = False)

plot_matrices(solution_matrix, problem_matrix)
```

# Increase difficulty (moderate to very advanced)
# `ruleset.moderate` -> `ruleset.very_advanced`

```python
solution_matrix, problem_matrix = create_matrix(ruleset.very_advanced, save = False)

plot_matrices(solution_matrix, problem_matrix)
```

# Add answer options (alternatives)

```python
solution_matrix, problem_matrix, alternatives = create_matrix(ruleset.very_advanced, alternatives=4, save = False)

plot_matrices(solution_matrix, problem_matrix, alternatives)
```

# Up to 16 alternatives

```python
solution_matrix, problem_matrix, alternatives = create_matrix(ruleset.very_advanced, alternatives=8, save = False)

plot_matrices(solution_matrix, problem_matrix, alternatives)
```

```python
solution_matrix, problem_matrix, alternatives = create_matrix(ruleset.very_advanced, alternatives=16, save = False)

plot_matrices(solution_matrix, problem_matrix, alternatives)
```

# Adding reproducibility with a seed

```python
solution_matrix, problem_matrix, alternatives = create_matrix(ruleset.moderate, seed=42, alternatives=4, save = False)

plot_matrices(solution_matrix, problem_matrix, alternatives)
```

# We can basically generate unlimited reasoning puzzles → at any difficulty level → up to 16 answer options → and use seeds for reproducibility.
